```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

plt.style.use('seaborn-v0_8-white')

d1 = np.load('../data/friedman1.npz')
X1, y1 = d1['X_train'], d1['y_train']
d2 = np.load('../data/friedman2.npz')
X2, y2 = d2['X_train'], d2['y_train']
d3 = np.load('../data/friedman3.npz')
X3, y3 = d3['X_train'], d3['y_train']

fig = plt.figure(figsize=(14, 10))
gs = gridspec.GridSpec(3, 5, hspace=0.55, wspace=0.25)

# Figure text
fig.text(0.5, 0.895,
         r'$f(\mathbf{x}) = 10\sin(\pi x_1 x_2) + 20(x_3-0.5)^2 + 10x_4 + 5x_5 + \mathcal{N}(0,\sigma)$', 
         ha='center', fontsize=10.5)

fig.text(0.5, 0.595,
         r'$f(\mathbf{x}) = \sqrt{x_0^2 + \left(x_1 x_2 - (x_1 x_3)^{-1}\right)^2} + \mathcal{N}(0,\sigma)$', 
         ha='center', fontsize=10.5)

fig.text(0.5, 0.295,
         r'$f(\mathbf{x}) = \arctan\!\left(\frac{x_1 x_2 - (x_1 x_3)^{-1}}{x_0}\right) + \mathcal{N}(0,\sigma)$', 
         ha='center', fontsize=10.5)

# Plotting loops
for i in range(5):     # Friedman 1
    ax = fig.add_subplot(gs[0, i])
    ax.scatter(X1[:, i], y1, s=8, color='#3A5FA0', alpha=0.35, linewidths=0)
    ax.set_xlabel(f'$x_{i+1}$', fontsize=10)
    if i == 0:
        ax.set_ylabel('$y$ (normalized)', fontsize=9)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

for i in range(4):     # Friedman 2
    ax = fig.add_subplot(gs[1, i])
    ax.scatter(X2[:, i], y2, s=8, color='#C0392B', alpha=0.35, linewidths=0)
    ax.set_xlabel(f'$x_{i}$', fontsize=10)     # 0-indexed for Friedman 2 and Friedman 3
    if i == 0:
        ax.set_ylabel('$y$ (normalized)', fontsize=9)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

for i in range(4):     # Friedman 3
    ax = fig.add_subplot(gs[2, i])
    ax.scatter(X3[:, i], y3, s=8, color='#1A7A4A', alpha=0.35, linewidths=0)
    ax.set_xlabel(f'$x_{i}$', fontsize=10)
    if i == 0:
        ax.set_ylabel('$y$ (normalized)', fontsize=9)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

plt.savefig('../figures/figure1_friedman_datasets.png', dpi=300, bbox_inches='tight')
plt.show()
plt.close()
```
